```python
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter

df=pd.read_csv("/content/drive/MyDrive/Colab Notebooks/496pareto.csv")
print(df)

df.index = df['skills/competencies']
df = df.sort_values(by='importance',ascending=False)

df["cumulativePercentage"] = df["importance"].cumsum()/df["importance"].sum()*100

df["cumulativePercentage"]=df["cumulativePercentage"].apply(lambda x: round(x,2))
print(df)

fig, ax = plt.subplots()
ax.bar(df.index, df["importance"], color="C0" )

ax.set_xticklabels( df['skills/competencies'], rotation=45)


ax2 = ax.twinx()

ax2.plot(df.index, df["cumulativePercentage"], color="C1", marker="D", ms=7, label='ax2')
ax2.yaxis.set_major_formatter(PercentFormatter())

ax.tick_params(axis="y", colors="C0")
ax2.tick_params(axis="y", colors="C1")

for index, v in df.iterrows():
  label = round(v['cumulativePercentage'],1)
  plt.annotate(label, xy=(v['skills/competencies'], v['cumulativePercentage']))

plt.show()

```
