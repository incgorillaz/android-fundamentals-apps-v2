```python
import matplotlib.pyplot as plt
import requests
import pandas as pd


myAPIKey = '1ca87828fc2a29c6ab6cb85f59468568'

stock = 'FB'
balanceStatements = requests.get(f"https://financialmodelingprep.com/api/v3/balance-sheet-statement/{stock}?apikey={myAPIKey}").json()

incomeStatements = requests.get(f'https://financialmodelingprep.com/api/v3/income-statement/{stock}?apikey={myAPIKey}').json()


print(balanceStatements[0])

twoYearBalanceStatment= balanceStatements[:2]
twoYearIncomeStatement= incomeStatements[:2]


profitability=[]
assetTurnOver=[]
capitalStructure=[]
returnOnEquity=[]

for i in range (0,2):
  profitabilityValue=twoYearIncomeStatement[i]['netIncome']/twoYearIncomeStatement[i]['revenue']
  profitability.append(profitabilityValue)
  assetTurnOverValue=twoYearIncomeStatement[i]['revenue']/twoYearBalanceStatment[i]['totalAssets']
  assetTurnOver.append(assetTurnOverValue)
  capitalStructureValue=twoYearBalanceStatment[i]['totalAssets']/twoYearBalanceStatment[i]['totalStockholdersEquity']
  capitalStructure.append(capitalStructureValue)
  returnOnEquityValue=profitabilityValue*assetTurnOverValue*capitalStructureValue
  returnOnEquity.append(returnOnEquityValue)

year=[i+1 for i in range (0,2)]
plt.plot(year,profitability)
plt.xlabel('Year ')
plt.ylabel('Profitability')
plt.title('Facebook')
plt.show()

year=[2021,2020]
plt.plot(year,assetTurnOver)
plt.xlabel('Year ')
plt.ylabel('AssetTurnOver')
plt.title('Facebook')
plt.show()

year=[2021,2020]
plt.plot(year,capitalStructure)
plt.xlabel('Year ')
plt.ylabel('CapitalStructure')
plt.title('Facebook')
plt.show()

year=[2021,2020]
plt.plot(year,returnOnEquity)
plt.xlabel('Year ')
plt.ylabel('ReturnOnEquityy')
plt.title('Facebook')
plt.show()

print(returnOnEquity)  


# Fb profitability, asset turnover, capital structure and return on equity grow year by year, making it a stock worth investing in
```
